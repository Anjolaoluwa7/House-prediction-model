```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline 
import matplotlib
matplotlib.rcParams["figure.figsize"]= (20,10)
```

```python
df7 = pd.read_csv("Bengaluru_House_Data.csv")
df7.head()
```

```python
df7.shape
```

```python
df6 = df7.drop(['area_type','society','balcony'],axis = 'columns')
df6.head()
```

```python
df6.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except :
        return False
    return True 
```

```python
df6[~df6['total_sqft'].apply(is_float)].head()
```

```python
def convert_total_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2 :
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df5 = df6.copy()
df5['total_sqft'] = df5['total_sqft'].apply(convert_total_sqft_to_num)
df5.head(3)
```

```python
df5["price_per_sqft"] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
df5 = df5.drop(["availability"],axis ='columns')
df5
```

```python
df5.location  = df5.location.apply(lambda x : x.strip() if isinstance(x, str) else x)
location_stats = df5.groupby("location")["location"].agg("count").sort_values(ascending = False)
location_stats
```

```python
 len(location_stats[location_stats <= 15])
```

```python
location_less_than_10 = location_stats[location_stats <= 15]
location_less_than_10
```

```python
df5.location  = df5.location.apply(lambda x : 'other' if  x in location_less_than_10 else x)
len(df5.location.unique())
```

```python
df5['no_of_rooms'] = df5['size'].apply(lambda x: int(x.split(' ')[0]) if isinstance(x, str) else x)
```

```python
df5.head()
```

```python
df6 = df5[~(df5.total_sqft/df5.no_of_rooms<300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)   
    return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def remove_no_of_rooms_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        no_of_rooms_stats = {}
        for no_of_rooms,no_of_rooms_df in location_df.groupby('no_of_rooms'):
            no_of_rooms_stats[no_of_rooms] = {
                     'mean': np.mean(no_of_rooms_df.price_per_sqft),
                     'std':np.std(no_of_rooms_df.price_per_sqft),
                     'count':no_of_rooms_df.shape[0]
                }
        for no_of_rooms,no_of_rooms_df in location_df.groupby("no_of_rooms"):
            stats = no_of_rooms_stats.get(no_of_rooms-1)
            if  stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices,no_of_rooms_df[no_of_rooms_df.price_per_sqft<(stats["mean"])].index.values)
    return df.drop(exclude_indices,axis='index')   

df8 = remove_no_of_rooms_outliers(df7)
df8.shape
```

```python
df9= df8[df8.bath<df8.no_of_rooms+2]
df9
```

```python
df10 = df9.drop(['price_per_sqft','size'], axis='columns')
df10
```

```python
dummies = pd.get_dummies(df10.location)
dummies
```

```python
df11= pd.concat([df10,dummies.drop('other', axis = 'columns')] , axis = 'columns')
df11
```

```python
df12 = df11.drop('location', axis = 'columns')
df12
```

```python
X = df12.drop('price', axis = 'columns')
X
```

```python
y = df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size = 0.3 ,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_f = LinearRegression()
lr_f.fit(X_train,y_train)
lr_f.score(X_test,y_test)
```

```python
def predict_price(location,sqft,bath,no_of_rooms):
    loc_index = np.where(X.columns==location)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = no_of_rooms
    if loc_index >= 0:
        x[loc_index] = 1
    
    return lr_f.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1394,2,2)
```
